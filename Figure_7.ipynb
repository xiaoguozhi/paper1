```python
# import pandas
import pandas as pd
# import matplotlib
import matplotlib.pyplot as plt
# import seaborn
import seaborn as sns
%matplotlib inline
```

```python
dat = pd.read_csv('D:/data/Electricity_data/pattern_forecasting/errdfnew.csv')
print(dat.head(3))
```

```python
# make grouped boxplot and save it in a variable
bp = sns.boxplot(y='Accuracy', x='Model', 
                 data=dat, 
                 palette="colorblind", 
                hue='Data')
bp = sns.stripplot(y='Accuracy', x='Model', 
                   data=dat, 
                   jitter=True,
                   dodge=True, 
                   marker='o', 
                   alpha=0.7,
                   hue='Data',
                   color='grey')
# get legend information from the plot object
handles, labels = bp.get_legend_handles_labels()
# specify just one legend
l = plt.legend(handles[0:2], labels[0:2])
# make grouped stripplot and save it in a variable
```
