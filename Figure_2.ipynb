```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#df=pd.read_csv('F:/2021-10-22/data/paper_data_and_code/Monthdf.csv')
#df=pd.read_csv('F:/2021-10-22/data/paper_data_and_code/Temperaturedf.csv')
df=pd.read_csv('F:/2021-10-22/data/paper_data_and_code/Weekdf.csv')
```

```python
df.tail()
```

```python
fig=plt.figure(figsize=(6,5));
ax = fig.add_subplot(1, 1, 1)
ax.plot(df['Temperature'], df['STemperature'],c='black')
ax.plot(df['Temperature'], df['lower'], c='gray', ls='--')
ax.plot(df['Temperature'], df['upper'], c='gray', ls='--')
ax.axhline(y=0,linewidth=1, color='r',linestyle='--')
ax.set_ylim(-5,10)
plt.xlabel('Temperature',fontsize=14)
plt.ylabel('s(Temperature)',fontsize=14)
plt.show()
```

```python
fig=plt.figure(figsize=(6,5));
ax = fig.add_subplot(1, 1, 1)
ax.plot(df['Week'], df['SWeek'],c='black')
ax.plot(df['Week'], df['lower'], c='gray', ls='--')
ax.plot(df['Week'], df['upper'], c='gray', ls='--')
ax.axhline(y=0,linewidth=1, color='r',linestyle='--')
ax.set_ylim(-2,2)
plt.xlabel('Day of the Week',fontsize=14)
plt.ylabel('s(Week)',fontsize=14)
plt.show()
```

```python
fig=plt.figure(figsize=(6,5));
ax = fig.add_subplot(1, 1, 1)
ax.plot(df['Month'], df['SMonth'],c='black')
ax.plot(df['Month'], df['lower'], c='gray', ls='--')
ax.plot(df['Month'], df['upper'], c='gray', ls='--')
ax.axhline(y=0,linewidth=1, color='r',linestyle='--')
ax.set_ylim(-5,6)
plt.xlabel('Month of the Year',fontsize=14)
plt.ylabel('s(Month)',fontsize=14)
plt.show()
```
